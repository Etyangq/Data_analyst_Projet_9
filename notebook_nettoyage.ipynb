### Nettoyage du fichier

#### Créer dataframe: dispo_alim_total

```python
# Verifier les valeurs uniques de la colonne 'Unité'
dispo_alim.Unité.unique()
```

```python
# Sélection des colonnes utiles
dispo_alim = dispo_alim[['Zone', 'Élément', 'Produit', 'Unité','Valeur']]
dispo_alim.head()
```

```python
# Changement de l'unité des valeurs en tonnes
dispo_alim['Valeur']*= 1e6
```

```python
#Pivot dataframe 'dispo_alim'
dispo_alim_total = dispo_alim.pivot_table(columns='Élément', index=['Zone'], values='Valeur',aggfunc=np.sum)
```

```python
dispo_alim_total.head()
```

```python
# Sélection des colonnes utiles
dispo_alim_total = dispo_alim_total[['Exportations - Quantité', 'Importations - Quantité', 'Nourriture']]
dispo_alim_total.head()
```

```python
# Renommage des colonnes
dispo_alim_total =dispo_alim_total.rename(columns={'Exportations - Quantité': 'Exportation totale', 'Importations - Quantité': 'Importations Totale', 'Nourriture': 'Consommation Totale'})
dispo_alim_total.head()
```

```python
dispo_alim_total.info()
```

#### Créer dataframe: dispo_poulet

```python
# Choisir uniquement les données des produits "Viande de Volailles"
dispo_poulet = dispo_alim[dispo_alim['Produit'] == 'Viande de Volailles'].reset_index(drop=True)
dispo_poulet.head()
```

```python
#Pivot dataframe 'dispo_poulet'
dispo_poulet = dispo_poulet.pivot(columns='Élément', index=['Zone'], values='Valeur')
```

```python
dispo_poulet.head()
```

***- Vérifier les valeurs nulles***

```python
dispo_poulet.isna().sum()
```

**Vu que j'ai besoin de la variable 'Exportations' qui a 37 valeurs nulles, alors je vais remplacer ces valeurs nulles par cette formule:**

Disponibilité Intérieure = (Production + Importations - disponibilité alimentaire(Quantité) + Variation de stock) - Exportations
devient alors:  
**Exportations = (Production + Importations - disponibilité alimentaire(Quantité) + Variation de stock) - Disponibilité Intérieure**

```python
dispo_poulet['Exportations - Quantité']= dispo_poulet.apply(lambda x : ((x['Production'] + x['Importations - Quantité'] 
                                                                       - x['Disponibilité alimentaire en quantité (kg/personne/an)'] 
                                                                       + x['Variation de stock']) - x['Disponibilité intérieure']) 
                                                          if pd.isnull(x['Exportations - Quantité']) else x['Exportations - Quantité'], axis=1)
```

```python
dispo_poulet.isna().sum()
```

J'ai remplacé 33 valeurs manquantes avec cette formule.

```python
# Sélection des colonnes utiles
dispo_poulet = dispo_poulet[['Disponibilité intérieure', 'Exportations - Quantité',
                           'Importations - Quantité', 'Production', 'Nourriture']]
dispo_poulet.head()
```

```python
# Changer les noms des colonnes
dispo_poulet = dispo_poulet.rename(columns={'Disponibilité intérieure': 'Dispo intérieure poulet', 
                        'Exportations - Quantité': 'Exportations poulet',
                        'Importations - Quantité': 'Importations poulet', 
                        'Production': 'Production poulet','Nourriture': 'Consommation poulet'})
dispo_poulet.head()
```

```python
dispo_poulet.info()
```

#### Jointure dispo_poulet et dispo_alim_total 

```python
dispo_poulet_17 = pd.merge(dispo_alim_total, dispo_poulet, how='inner', on='Zone')
```

```python
dispo_poulet_17 = dispo_poulet_17.reset_index()
```

```python
dispo_poulet_17.head()
```

#### Ajouter les variables ####  
- La proportion de poulet par rapport à l'exportation total, l'importation totale et la consommation totale

```python
dispo_poulet_17['Taux_import_poulet %'] = round((dispo_poulet_17['Importations poulet']*100/ dispo_poulet_17['Importations Totale']),2)
```

```python
dispo_poulet_17['Taux_expot_poulet %'] = round((dispo_poulet_17['Exportations poulet']*100 / dispo_poulet_17['Exportation totale']),2)
```

```python
dispo_poulet_17['Taux_consom_poulet %'] = round((dispo_poulet_17['Consommation poulet']*100/ dispo_poulet_17['Consommation Totale']),2)
```

```python
dispo_poulet_17.head()
```

```python
# Suppression des colonnes inutiles
dispo_poulet_17 = dispo_poulet_17.drop(
    columns=['Exportation totale','Importations Totale', 'Consommation Totale'])
```

```python
dispo_poulet_17.head()
```
